```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.randn(1000)
width=.5
histmin=np.floor(min(x))
histmax=np.ceil(max(x))+width
bins=np.arange(histmin,histmax,width)
plt.hist(x,bins=100,alpha=.5,edgecolor='black')
plt.ylabel('y')
plt.xlabel('x')
plt.savefig('lotsofbins.pdf',bbox_inches='tight')
```
